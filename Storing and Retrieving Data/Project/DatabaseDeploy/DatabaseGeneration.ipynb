```python
from faker import Faker
import numpy as np
import pandas as pd
import random
```

```python
fake = Faker('pt_pt')
```

```python
DBName = "ars_project_group_d"
```

```python
ClientSize = 200
ClientCategorySize = 4
TaxSize = 2
ProductsSize = 200
InvoiceSize = 3000
```

# Clients

```python
_id = np.arange(1,ClientSize+1,1)
_category_id = np.random.randint(1,ClientCategorySize+1, size=ClientSize)
_tax_id = np.random.randint(1,TaxSize+1, size=ClientSize)
_first_name = [fake.first_name() for _ in range(ClientSize)]
_last_name = [fake.last_name() for _ in range(ClientSize)]
_age = [fake.pyint(min_value=1, max_value=100, step=1) for _ in range(ClientSize)]
_phone_number = [fake.phone_number() for _ in range(ClientSize)]
_email = [f"{person[0]}_{person[1]}_{person[2]}@SuperDuperStore.com" for person in zip(_first_name,_last_name,_age)]
_street_address = [fake.street_address() for _ in range(ClientSize)]
_city = [fake.city() for _ in range(ClientSize)]
_state = [fake.concelho() for _ in range(ClientSize)]
_country = [fake.country() for _ in range(ClientSize)]
_zip_code = [fake.postcode() for _ in range(ClientSize)]
_vat = [fake.vat_id() for _ in range(ClientSize)]

ClientData = zip(
                _id,
                _category_id,
                _tax_id,
                _first_name,
                _last_name,
                _age,
                _phone_number,
                _email,
                _street_address,
                _city,
                _state,
                _country,
                _zip_code,
                _vat
                )

ClientsColumns =   ['id',
                    'category_id',
                    'tax_id',
                    'first_name',
                    'last_name',
                    'age',
                    'phone_number',
                    'email',
                    'street_address',
                    'city',
                    'state',
                    'country',
                    'zip_code',
                    'vat']

CLIENT = pd.DataFrame(ClientData, columns = ClientsColumns)
CLIENT.head()
```

```python
with open('Faker_Client.sql','w', encoding='utf8') as f:
    for row in CLIENT.itertuples():
        sadrress = str(row.street_address).replace("'",'')
        f.write(f"INSERT INTO `clients` (`id`, `category_id`, `tax_id`, `first_name`, `last_name`, `age`, `phone_number`, `email`, `street_address`, `city`, `state`, `country`, `zip_code`, `vat_number`) VALUES ({row.id}, {row.category_id}, {row.tax_id}, '{row.first_name}', '{row.last_name}', {row.age}, '{row.phone_number}', '{row.email}', '{sadrress}', '{row.city}', '{row.state}', '{row.country}', '{row.zip_code}', '{row.vat}');\n")
```

# Products

```python
type_ = ['Service','Product']
segment_ = ['House Cleaners','Detergent','Air Care','Fabric Softener','Moisturizing','Desinfectant']

_id = np.arange(1,ProductsSize+1,1)
_name = [fake.ean() for _ in range(ProductsSize)]
_description = [None for _ in range(ProductsSize)]
_type = [ type_[random.randint(0,len(type_)-1)] for _ in range(ProductsSize)]
_segment = [ segment_[random.randint(0,len(segment_)-1)] for _ in range(ProductsSize)]
_brand = [fake.last_name() for _ in range(ProductsSize)]
_price = [float(fake.pydecimal(left_digits=4, right_digits=2, positive=True, min_value=0, max_value=1000))
           for _ in range(ProductsSize)]
_stock = [fake.pyint() for _ in range(ProductsSize)]
_avg_rating_FULL = [float(fake.pydecimal(left_digits=1, right_digits=1, positive=True, min_value=0, max_value=5))
               for _ in range(ProductsSize)]
_avg_rating = [ r if random.randint(0,100) < 90 else None for r in _avg_rating_FULL ] # introduce NULLs for G-5

ProductsData = zip(
                _id,
                _name,
                _description,
                _type,
                _segment,
                _brand,
                _price,
                _stock,
                _avg_rating
                )

ProductsColumns =   ['id',
                    'name',
                    'description',
                    'type',
                    'segment',
                    'brand',
                    'price',
                    'stock',
                    'avg_rating']

PRODUCT = pd.DataFrame(ProductsData, columns = ProductsColumns).fillna('NULL')
PRODUCT.head(100)
```

```python
with open('Faker_Product.sql','w', encoding='utf8') as f:
    for row in PRODUCT.itertuples():
        f.write(f"INSERT INTO `products` (`id`,`name`,`description`,`type`,`segment`,`brand`,`price`,`stock`,`avg_rating`) VALUES ({row.id}, '{row.name}', '{row.description}', '{row.type}', '{row.segment}', '{row.brand}', {round(row.price,2)}, {row.stock}, {row.avg_rating});\n")
```

# Invoice Details

```python
# Generate sequential index and then randomly duplicate invoices (multiple products per invoice)
_invoice_id = np.concatenate((np.arange(1,InvoiceSize+1,1),np.random.randint(1,InvoiceSize+1, size=InvoiceSize)))
_product_id = np.random.randint(1,ProductsSize+1, size=len(_invoice_id))
_quantity = [fake.pyint(min_value=1, max_value=20, step=1) for _ in range(len(_invoice_id))]


invoice_details = zip(
                    _invoice_id,
                    _product_id,
                    _quantity
                    )

InvoiceDetailsColumns =   [ 'invoice_id',
                            'product_id',
                            'quantity']

ID = pd.DataFrame(invoice_details, columns = InvoiceDetailsColumns)
ID = ID.drop_duplicates(subset=['invoice_id','product_id'], keep='first')
ID.head()
```

```python
FULL = pd.merge(ID, PRODUCT, how='inner', left_on='product_id', right_on='id')

FULL['Amount'] = FULL['quantity'] * FULL['price']

FULL.head(10)
```

```python
with open('Faker_InvoiceDetails.sql','w', encoding='utf8') as f:
    for row in FULL.itertuples():
        f.write(f"INSERT INTO `invoice_details` (`invoice_id`,`product_id`,`quantity`,`Amount`) VALUES ({row.invoice_id}, '{row.product_id}', '{row.quantity}', '{round(row.Amount,2)}');\n")
```

# Invoices

```python
Invoices = FULL.groupby('invoice_id',as_index = False).Amount.sum()
Invoices['client_id'] = np.random.randint(1,ClientSize+1, size=Invoices.shape[0])
Invoices['invoice_status_id'] = np.random.randint(1,4, size=Invoices.shape[0])

Invoices = pd.merge(Invoices, CLIENT, how='inner', left_on='client_id', right_on='id')

Invoices = pd.merge(Invoices, TAX, how='inner', left_on='tax_id', right_on='id')

Invoices['Discount'] = 0
Invoices['company_id'] = 1
Invoices['invoice_date'] = [fake.date_between(start_date='-3y', end_date='today').strftime('%Y-%m-%d') 
                            for _ in range(Invoices.shape[0])]

Invoices['payment_due_date'] = pd.to_datetime(Invoices['invoice_date']) + pd.DateOffset(days=2)

Invoices['tax_amount'] = Invoices['Amount'] * (Invoices['rate'] / 100.0)
Invoices['total'] = Invoices['Amount'] - Invoices['Discount'] + Invoices['tax_amount']


Invoices = Invoices[['invoice_id','client_id','company_id','invoice_status_id','invoice_date','payment_due_date','Amount','Discount','rate','tax_amount','total']]

Invoices.head()
```

```python
with open('Faker_Invoices.sql','w', encoding='utf8') as f:
    for row in Invoices.itertuples():
        f.write(f"INSERT INTO `invoices` (`id`,`client_id`,`company_id`,`invoice_status_id`,`invoice_date`,`payment_due_date`,`subtotal`,`discount`,`tax_rate`,`tax_amount`,`total`) VALUES ({row.invoice_id}, {row.client_id}, {row.company_id}, {row.invoice_status_id}, '{row.invoice_date}','{row.payment_due_date}', {round(row.Amount,2)}, {round(row.Discount,2)}, {row.rate}, {round(row.tax_amount,2)}, {round(row.total,2)});\n")
```
